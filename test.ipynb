````python
from pymongo import MongoClient

# client = MongoClient()

# db = client("")

client = MongoClient("mongodb://localhost:27017/")

db = client["AuteursDB"]
collection = db["Auteurs"]
````

````python
authors = [
    {
        "Name": "Victor Hugo",
        "age": 45,
        "Novels": ["Notre-Dame de Paris", "Les Misérables", "Quatrevingt-treize"],
        "Poems": [
            {"year": 1822, "title": "Odes et poésies diverses"},
            {"year": 1825, "title": "Nouvelles Odes"},
            {"year": 1828, "title": "Odes et Ballades"},
            {"year": 1834, "title": "Claude Gueux"}
        ]
    },
    {
        "Name": "Ernest Miller Hemingway",
        "Novels": ["The Torrents of Spring", "The Sun Also Rises", "A Farewell to Arms"]
    },
    {
        "Name": "Gabriel García Márquez",
        "Novels": ["El coronel no tiene quien le escriba", "Crónica de una muerte anunciada", "Cien años de soledad"],
        "ShortStories": [
            {"year": 1962, "title": "Los funerales de la Mamá Grande"},
            {"year": 1962, "title": "Ojos de perro azul"}
        ]
    },
    {"Name": "T1", "status": "Test", "Novels": ["RT1", "RT2"], "Poems": [{"year":21, "title":"PT1"}, {"year":22, "title":"PT2"}]},
    {"Name": "T2", "status": "Test", "Novels": ["RS1", "RS2"], "Poems": [{"year":21, "title":"PS1"}, {"year":22, "title":"PS2"}]},
    {"Name": "T3", "status": "Test", "Novels": ["R1", "R2"], "Poems": [{"year":21, "title":"P1"}, {"year":22, "title":"P2"}]}
]


collection.insert_many(authors)
````

Update Victor Hugo’s age to 46 → update_one().

````python
collection.update_one({"Name" : 'Victor Hugo' } , {"$set" : {"Name" : "test_update"}})
````

Add age 56 to Ernest Miller Hemingway → update_one().

````python
collection.update_one({"Name" : "Ernest Miller Hemingway" } , {"$set" : {"age" : 99}} , True)
````

Rename fields "Nom" → "Name" and "pays" → "country" for Victor Hugo and Ernest Hemingway → update_many().

````python
collection.update_many({"Name" : {"$in" : ['Victor Hugo' , "Ernest Miller Hemingway"]} } , {"$rename" : {"Name" : "nom" , "pays" : "country"}} , True)
````

Remove the fields "age" and "country" from Victor Hugo → update_one().

````python
collection.update_one({"nom" : "Ernest Miller Hemingway" } , {"$unset" : {"age" : ""}})
````

Delete author T1 → delete_one().

````python
collection.delete_one({"Name" : "T1" })
````

Delete all authors with "status": "Test" → delete_many().

````python
collection.delete_many({"status" : "Test" })
````
